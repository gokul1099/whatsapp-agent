```python
from langgraph.checkpoint.sqlite.aio import AsyncSqliteSaver
from langchain_core.messages import AIMessageChunk, HumanMessage
from ai_companion.settings import settings
from ai_companion.graph.graph import graph_builder
from ai_companion.graph.constants import NodeName
from dotenv import load_dotenv
load_dotenv()
```

```python
user_message = "I am gokul"
thread_id = 1
async with AsyncSqliteSaver.from_conn_string(settings.SHORT_TERM_MEMORY_DB_PATH) as short_term_memory:
    graph = graph_builder.compile()
    async for chunk in graph.astream(
        {"messages": [HumanMessage(content=user_message)]},
        {"configurable": {"thread_id": thread_id}},
        stream_mode = "messages"    
    ):
    
        msg, metadata = chunk
        if metadata.get("langgraph_node") == NodeName.CONVERSATION and isinstance(msg, AIMessageChunk):
            print(msg)
```

```python
##Image node
user_message = "My name is gokul"
thread_id = 1
async with AsyncSqliteSaver.from_conn_string(settings.SHORT_TERM_MEMORY_DB_PATH) as short_term_memory:
    graph = graph_builder.compile()
    async for chunk in graph.astream(
        {"messages": [HumanMessage(content=user_message)]},
        {"configurable": {"thread_id": thread_id}},
        stream_mode = "updates"    
    ):
        print("TURN_1",chunk)
    print("***********************************")
    print("***********************************")
    print("***********************************")
     # Turn 2 - Context from Turn 1 is automatically loaded
    async for chunk in graph.astream(
        {"messages": [HumanMessage(content="I am a software engineer")]},  # ← Just new message
        {"configurable": {"thread_id": 1}},  # ← Same thread_id
        stream_mode="updates"
    ):
        print("TURN_2",chunk)
    print("***********************************")
    print("***********************************")
    print("***********************************")

    async for chunk in graph.astream(
        {"messages": [HumanMessage(content="I am currently learning software engineer")]},  # ← Just new message
        {"configurable": {"thread_id": 1}},  # ← Same thread_id
        stream_mode="updates"
    ):
        print("TURN_3",chunk)
    print("***********************************")
    print("***********************************")
    print("***********************************")
```
